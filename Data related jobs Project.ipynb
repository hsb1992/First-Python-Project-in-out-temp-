```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seaborn
```

```python
df = pd.read_csv('Uncleaned_DS_jobs.csv')
```

```python
#changing all -1 and unknowns to na for later use
object_col = df.select_dtypes(include = 'object').columns
```

```python
df = df[object_col].replace('-1','na')
```

```python
df = df[object_col].replace('Unknown','na')
```

# Salary Estimate Cleaning

```python
df['Salary Estimate'] = df['Salary Estimate'].str.split('(').str[0].str.strip()
```

```python
df['min_salary'] = df['Salary Estimate'].str.split('-').str[0].str.strip()
```

```python
df['max_salary'] = df['Salary Estimate'].str.split('-').str[1].str.strip()
```

```python
df['min_salary'] = df['min_salary'].str.replace('$','')
df['max_salary'] = df['max_salary'].str.replace('$','')
```

```python
df['min_salary']= df['min_salary'].str.replace('K','').astype('int')*1000
```

```python
df['max_salary']= df['max_salary'].str.replace('K','').astype('int')*1000
```

```python
df[['Salary Estimate','max_salary','min_salary']].tail()
```

# Job Description cleaning

```python
df['excel'] = df['Job Description'].map(lambda x : 1 if 'excel' in x.lower() else 0)
df['python'] = df['Job Description'].map(lambda x : 1 if 'python' in x.lower() else 0)
df['hadoop'] = df['Job Description'].map(lambda x : 1 if 'hadoop' in x.lower() else 0)
df['big data'] = df['Job Description'].map(lambda x : 1 if 'big data' in x.lower() else 0)
df['tableau'] = df['Job Description'].map(lambda x : 1 if 'tableau' in x.lower() else 0)
df['spark'] = df['Job Description'].map(lambda x : 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].map(lambda x : 1 if 'aws' in x.lower() else 0)
```
